```python
import numpy as np
from matplotlib.pyplot import imread, imshow
import os
from tqdm import tqdm
```

```python
directory = './data/'
files = os.listdir(directory)
```

```python
blacklist = []
for filename in tqdm(files):
    image = imread(directory+filename)
    if len(image.shape)!=3: blacklist.append(filename)
```

```python
correct_files = list(set(files)-set(blacklist))
```

```python
numpy_images = []
for filename in tqdm(correct_files):
    numpy_images.append(np.expand_dims(imread(directory+filename),0))
    
dataset = np.concatenate(numpy_images)
```
